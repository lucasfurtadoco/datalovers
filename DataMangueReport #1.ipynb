# Data Visualization (Visualiza√ß√£o dos Dados)

Shoooow! Nosso conjunto de dados est√° limpo e tratado agora podemos gerar <i>insights</i> sobre eles!

A primeira pergunta que eu quero responder √©:

<b>Qual foi a despesa total no ano de 2017 da Prefeitura do Recife?</b>

````python
print('O valor total de despesas entre todas as pastas foi de R$ {:,.2f}.'.format(df['VALOR_PAGO'].sum()))
````

<b>Quais s√£o as pastas subordinadas a Prefeitura e qual delas que realizou maior quantidade de empenhos?</b>

Para responder a esse questionamento, criei a vari√°vel ```count_orgao```, que ir√° armazenar as informa√ß√µes desse conjunto de dados agrupado por org√£o.

````python
count_orgao = df.groupby('ORGAO')['CONTAGEM'].sum().sort_values().plot(kind='barh', color=random.choice(cores), grid=True, figsize=(13,13), alpha=.8)

for p in count_orgao.patches:
    b=p.get_bbox()
    count_orgao.annotate("{:,.0f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR ORG√ÉO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('QUANTIDADE DE EMPENHOS')
plt.show()
````

No gr√°fico acima, podemos ver que a Prefeitura do Recife possui <u>39 pastas</u>.<br>
E no ano de <b>2017</b> a pasta que mais emitiu notas de empenho foi a <b>SECRETARIA DE SA√öDE</b>, com um total de <b>15,583 empenhos</b>.

Ap√≥s classificarmos as pastas por empenhos... A pergunta √©:

<b>Ser√° que a secretaria que mais emitiu notas de empenho foi a que mais gastou dinheiro?</b>

````python
sum_orgao_pago = df.groupby('ORGAO')['VALOR_PAGO'].sum().sort_values().plot(kind='barh', color=random.choice(cores), figsize=(13,13), grid=True, alpha=.8)

for p in sum_orgao_pago.patches:
    b=p.get_bbox()
    sum_orgao_pago.annotate("{:,.2f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR ORG√ÉO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('VALOR PAGO (em R$)')
plt.show()
````

Podemos observar que a <b>SECRETARIA DE SA√öDE</b> se mant√©m NO TOPO das despesas da prefeitura chegando a custar pouco mais que 1 bilh√£o de reais para os cofres p√∫blicos.

Em seguida temos a <b>SECRETARIA DE EDUCA√á√ÉO</b> em segundo lugar, a <b>SECRETARIA DE INFRAESTRUTURA E SERVI√áOS URBANOS</b> em terceiro lugar, a <b>SECRETARIA DE ADMINISTRA√á√ÉO E GEST√ÉO DE PESSOAS</b> em quarto lugar e os <b>ENCARGOS GERAIS DO MUNICIPIO</b> finalizando as TOP 5 despesas da Prefeitura do Recife, essas 5 pastas correspondem a <u>61,31%</u> de tudo que foi pago no ano de 2017.

<b> E como foi a distribui√ß√£o dos empenhos durante os meses do ano?</b>

````python
df.groupby('MES')['CONTAGEM'].sum().plot(color='purple', figsize=(12,5), grid=True)

plt.title('QUANTIDADE DE EMPENHOS AO LONGO DO ANO \n Total: ' + str(df['CONTAGEM'].sum()) + ' ocorr√™ncias')
plt.xlabel('MESES')
plt.ylabel('QUANTIDADE DE EMPENHOS')
plt.show()
````

No gr√°fico plotado dos empenhos de 2017, observamos que exite um patamar praticamente est√°vel durante todo o ano, mas apresenta um grande pico no m√™s de <b>DEZEMBRO</b>.

<b> Como √© a rela√ß√£o MENSAL entre os valores empenhados, liquidados e o realmente pago? </b>

````python
df.groupby('MES')['VALOR_EMPENHADO','VALOR_LIQUIDADO','VALOR_PAGO'].sum().plot(figsize=(12,5), grid=True)

plt.title('VALORES EMPENHADOS, LIQUIDADOS E PAGOS AO LONGO DO ANO \n Total: ' + str(df['CONTAGEM'].sum()) + ' ocorr√™ncias')
plt.xlabel('MESES')
plt.ylabel('VALOR (em R$)')
plt.show()
````

Segundo o Portal da Transpar√™ncia do MS,

O <b>empenho</b> √© o valor que o org√£o p√∫blico reserva para efetuar um pagamento planejado, podendo ocorrer ap√≥s a assinatura de um contrato de presta√ß√£o de servi√ßo por exemplo. Quando cada servi√ßo for executado o valor √© <b>liquidado</b>, e quando o prestador de servi√ßo receber o valor do servi√ßo conclu√≠do este √© considerado o valor <b>pago</b>.

<b> E como ficou no final do ano os valores empenhados, liquidados e pagos?</b>


````python
print('Valor empenhado: R$ {:,.2f}.'.format(df['VALOR_EMPENHADO'].sum()))
print('Valor liquidado: R$ {:,.2f}.'.format(df['VALOR_LIQUIDADO'].sum()))
print('Valor pago: R$ {:,.2f}.'.format(df['VALOR_PAGO'].sum()))

em_aberto = df['VALOR_EMPENHADO'].sum() - df['VALOR_PAGO'].sum()
print('Valor em aberto aguardando pagamento: R$ {:,.2f}.'.format(em_aberto))
````

<b> Quais s√£o as modalidades de pagamento utilizadas pela PCR?</b>

````python
count_licitacao = df.groupby('MODALIDADE_LICITACAO')['CONTAGEM'].sum().sort_values().plot(kind='barh', color=random.choice(cores), figsize=(13,13), grid=True,alpha=.8)

for p in count_licitacao.patches:
    b=p.get_bbox()
    count_licitacao.annotate("{:,.0f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR MODALIDADES DE PAGAMENTO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('MODALIDADES DE PAGAMENTO')
plt.xlabel('QUANTIDADE DE EMPENHOS')
plt.show()
````

Nessa gr√°fico a gente observa que a grande maioria dos pagamentos foram dispensados e/ou n√£o informados. <u>Algo que ao meu ver √© preocupante.</u>

````python
count_modalidade_disp = df[df['MODALIDADE_LICITACAO'] == "DISPENSADO"]
porcentagem_modalidade_disp = (count_modalidade_disp.count()/df.count())*100

print('{:.2f}% dos empenhos foram dispensados de concorr√™ncia.'.format(porcentagem_modalidade_disp[0]))
````

````python
count_modalidade_nao_info = df[df['MODALIDADE_LICITACAO'] == "N√ÉO INFORMADA"]
porcentagem_modalidade_nao_info = (count_modalidade_nao_info.count()/df.count())*100

print('{:.2f}% dos empenhos n√£o foram informados a modalidade de pagamento.'.format(porcentagem_modalidade_nao_info[0]))
````

````python
sec_mod_lici = df[df['MODALIDADE_LICITACAO'] == "N√ÉO INFORMADA"]

sec_mod_lici_count_nao_info = sec_mod_lici.groupby('ORGAO')['VALOR_PAGO'].sum().sort_values().plot(kind='barh', color=random.choice(cores), grid=True, figsize=(13,13), alpha=.8)

for p in sec_mod_lici_count_nao_info.patches:
    b=p.get_bbox()
    sec_mod_lici_count_nao_info.annotate("{:,.2f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('VALOR GASTO POR SECRETARIA SEM INFORMAR A MODALIDADE DE PAGAMENTO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('VALOR PAGO (em R$)')
plt.show()
````
