# Importando as Bibliotecas

```python
import pandas as pd # Usado para criarmos DataFrames
import numpy as np # Usado para realizarmos c√°lculos matem√°ticos com melhor performance
import matplotlib.pyplot as plt # Usado para plotarmos gr√°ficos
import random # Usado para gerar algo aleatoriamente
import benford as bf # Usado para realizar o estudo pela Lei de Benford

%matplotlib inline
```

```python
# Defini√ß√£o das cores que iremos utilizar nos gr√°ficos na etapa de visualiza√ß√£o dos dados
cores  = [['#9c863a','#cfa81f','#efad3b','#d2952a','#e69a17'], # Mostarda
          ['#6a188c','#8d38b0','#9109c9','#873fa5','#82419d'], # Roxo
          ['#41873c','#2c6c27','#4d6f4b','#4f9a49','#209317'], # Verde
          ['#186fe6','#2166c4','#0e64d8','#327de4','#0d5cc9'], # Azul
          ['#2c8d7c','#1c7565','#50887e','#38c2aa','#1a8471']  # Ciano
         ]
```

# Leitura do Conjunto de Dados

```python
# Iniciando a leitura do arquivo
df = pd.read_csv('/Users/lucasfeitosa/Downloads/Prefeitura/Despesas/recife-dados-despesas-2017.csv', sep = ";")
```

# Data Wrangling (Limpeza de Dados)

```python
# Exibindo o dataframe como foi disponibilizado pelo Portal de Dados Abertos da Cidade do Recife
df.head(3)
```

```python
# Verificando como os dados est√£o sendo reconhecidos pelo Pandas.
df.dtypes
```

```python
# Listando todas as colunas disponiveis no dataframe
df.columns
```

```python
# Ocultando colunas que n√£o ser√£o utilizada na an√°lise
df.drop(['orgao_codigo','categoria_economica_codigo','grupo_despesa_codigo','modalidade_aplicacao_codigo',
         'elemento_codigo','subelemento_codigo','funcao_codigo','subfuncao_codigo','programa_codigo',
         'acao_codigo','fonte_recurso_codigo','empenho_modalidade_codigo','subempenho',
         'indicador_subempenho','credor_codigo','modalidade_licitacao_codigo','unidade_codigo',
         'unidade_nome','empenho_ano'], axis=1, inplace=True);
```

```python
df.shape
```

```python
df.columns = ['ANO','MES','ORGAO','CATEGORIA','GRUPO','MODALIDADE','ELEMENTO','SUBELEMENTO','FUNCAO',
              'SUBFUNCAO','PROGRAMA','ACAO','RECURSO','MODALIDADE_EMPENHO','N_EMPENHO','CREDOR','MODALIDADE_LICITACAO',
             'VALOR_EMPENHADO','VALOR_LIQUIDADO','VALOR_PAGO']
```

```python
# Substituindo a virgula por ponto
df['VALOR_PAGO'] = df['VALOR_PAGO'].apply(lambda x: x.replace(',','.'));
df['VALOR_EMPENHADO'] = df['VALOR_EMPENHADO'].apply(lambda x: x.replace(',','.'));
df['VALOR_LIQUIDADO'] = df['VALOR_LIQUIDADO'].apply(lambda x: x.replace(',','.'));
```

```python
# Modificando o tipo de formata√ß√£o da coluna
df['VALOR_PAGO'] = df['VALOR_PAGO'].astype(np.float64)
df['VALOR_EMPENHADO'] = df['VALOR_EMPENHADO'].astype(np.float64)
df['VALOR_LIQUIDADO'] = df['VALOR_LIQUIDADO'].astype(np.float64)
```

```python
# Inserindo uma nova coluna para realizar contagem
df['CONTAGEM'] = 1
```

```python
# Mostrando o dataframe ap√≥s a limpeza e tratamento
df.head()
```

# Data Visualization (Visualiza√ß√£o dos Dados)

Shoooow! Agora que o nosso conjunto de dados est√° limpo e tratado agora podemos gerar <i>insights</i> sobre eles!

A primeira pergunta que eu quero responder √©:

<h2>Qual foi a despesa total no ano de 2017 da Prefeitura do Recife?</h2>

```python
print('O valor total de despesas entre todas as pastas foi de R$ {:,.2f}.'.format(df['VALOR_PAGO'].sum()))
```

<h2>Quais s√£o as pastas subordinadas a Prefeitura e qual delas que realizou maior quantidade de empenhos?</h2>

```python
count_orgao = df.groupby('ORGAO')['CONTAGEM'].sum().sort_values().plot(kind='barh', color=random.choice(cores), grid=True, figsize=(13,13), alpha=.8)

for p in count_orgao.patches:
    b=p.get_bbox()
    count_orgao.annotate("{:,.0f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR ORG√ÉO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('QUANTIDADE DE EMPENHOS')
plt.show()
```

Com base no conjunto de dados disponibilizado pela Prefeitura, podemos observar que a Prefeitura do Recife possui <u>39 pastas</u>.<br>
E no ano de <b>2017</b> a pasta que mais emitiu notas de empenho foi a <b>SECRETARIA DE SA√öDE</b>, com um total de <b>15,583 empenhos</b>.

Voc√™ sabia disso?

Ap√≥s classificarmos as pastas por empenhos... A pergunta √©:

<h2>Ser√° que a secretaria que mais emitiu notas de empenho foi a que mais gastou dinheiro?</h2>

```python
sum_orgao_pago = df.groupby('ORGAO')['VALOR_PAGO'].sum().sort_values().plot(kind='barh', color=random.choice(cores), figsize=(13,13), grid=True, alpha=.8)

for p in sum_orgao_pago.patches:
    b=p.get_bbox()
    sum_orgao_pago.annotate("{:,.2f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR ORG√ÉO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('VALOR PAGO (em R$)')
plt.show()
```

Wow! Observamos que a <b>SECRETARIA DE SA√öDE</b> ainda se mant√©m NO TOPO das despesas da prefeitura chegando a custar pouco mais que <b>1 bilh√£o de reais</b> para os cofres p√∫blicos. üò≤

Em seguida temos a:<br>
2) SECRETARIA DE EDUCA√á√ÉO<br>
3) SECRETARIA DE INFRAESTRUTURA E SERVI√áOS URBANOS<br>
4) SECRETARIA DE ADMINISTRA√á√ÉO E GEST√ÉO DE PESSOAS<br>
5) ENCARGOS GERAIS DO MUNICIPIO<br>

Essas 5 pastas correspondem a <u>61,31%</u> de tudo que foi pago no ano de 2017.

<h2>E como foi a distribui√ß√£o dos empenhos durante os meses do ano?</h2>

```python
df.groupby('MES')['CONTAGEM'].sum().plot(color='purple', figsize=(12,5), grid=True)

plt.title('QUANTIDADE DE EMPENHOS AO LONGO DO ANO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.xlabel('MESES')
plt.ylabel('QUANTIDADE DE EMPENHOS')
plt.show()
```

No gr√°fico acima, podemos observar que exite um patamar praticamente est√°vel durante todo o ano, mas apresenta um grande pico no m√™s de <b>DEZEMBRO</b>.

<h2> Como √© a rela√ß√£o MENSAL entre os valores empenhados, liquidados e o realmente pago? </h2>

```python
df.groupby('MES')['VALOR_EMPENHADO','VALOR_LIQUIDADO','VALOR_PAGO'].sum().plot(figsize=(12,5), grid=True)

plt.title('VALORES EMPENHADOS, LIQUIDADOS E PAGOS AO LONGO DO ANO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.xlabel('MESES')
plt.ylabel('VALOR (em R$)')
plt.show()
```

Segundo o Portal da Transpar√™ncia do MS,

O <b>empenho</b> √© o valor que o org√£o p√∫blico reserva para efetuar um pagamento planejado, podendo ocorrer ap√≥s a assinatura de um contrato de presta√ß√£o de servi√ßo por exemplo. Quando cada servi√ßo for executado o valor √© <b>liquidado</b>, e quando o prestador de servi√ßo receber o valor do servi√ßo conclu√≠do este √© considerado o valor <b>pago</b>.

<h2> E como ficou no final do ano os valores empenhados, liquidados e pagos?</h2>


```python
print('Valor empenhado: R$ {:,.2f}.'.format(df['VALOR_EMPENHADO'].sum()))
print('Valor liquidado: R$ {:,.2f}.'.format(df['VALOR_LIQUIDADO'].sum()))
print('Valor pago: R$ {:,.2f}.'.format(df['VALOR_PAGO'].sum()))

em_aberto = df['VALOR_EMPENHADO'].sum() - df['VALOR_PAGO'].sum()
print('Valor em aberto aguardando pagamento: R$ {:,.2f}.'.format(em_aberto))
```

<h2> Quais s√£o as modalidades de pagamento utilizadas pela PCR?</h2>

```python
count_licitacao = df.groupby('MODALIDADE_LICITACAO')['CONTAGEM'].sum().sort_values().plot(kind='barh', color=random.choice(cores), figsize=(13,13), grid=True,alpha=.8)

for p in count_licitacao.patches:
    b=p.get_bbox()
    count_licitacao.annotate("{:,.0f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('QUANTIDADE DE EMPENHOS AGRUPADOS POR MODALIDADES DE PAGAMENTO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('MODALIDADES DE PAGAMENTO')
plt.xlabel('QUANTIDADE DE EMPENHOS')
plt.show()
```

Nesse gr√°fico a gente observa que a grande maioria dos pagamentos foram dispensados e/ou n√£o informados. <u>Algo que ao meu ver √© preocupante.</u>

```python
count_modalidade_disp = df[df['MODALIDADE_LICITACAO'] == "DISPENSADO"]
porcentagem_modalidade_disp = (count_modalidade_disp.count()/df.count())*100

print('{:.2f}% dos empenhos foram dispensados de concorr√™ncia.'.format(porcentagem_modalidade_disp[0]))
```

```python
count_modalidade_nao_info = df[df['MODALIDADE_LICITACAO'] == "N√ÉO INFORMADA"]
porcentagem_modalidade_nao_info = (count_modalidade_nao_info.count()/df.count())*100

print('{:.2f}% dos empenhos n√£o foram informados a modalidade de pagamento.'.format(porcentagem_modalidade_nao_info[0]))
```

```python
sec_mod_lici = df[df['MODALIDADE_LICITACAO'] == "N√ÉO INFORMADA"]

sec_mod_lici_count_nao_info = sec_mod_lici.groupby('ORGAO')['VALOR_PAGO'].sum().sort_values().plot(kind='barh', color=random.choice(cores), grid=True, figsize=(13,13), alpha=.8)

for p in sec_mod_lici_count_nao_info.patches:
    b=p.get_bbox()
    sec_mod_lici_count_nao_info.annotate("{:,.2f}".format(b.x1 + b.x0), (p.get_x() + p.get_width() + 1, p.get_y() - 0.05))

plt.title('VALOR GASTO POR SECRETARIA SEM INFORMAR A MODALIDADE DE PAGAMENTO \n Total: ' + str(df['CONTAGEM'].sum()) + ' empenhos')
plt.ylabel('ORG√ÉO')
plt.xlabel('VALOR PAGO (em R$)')
plt.show()
```

# Especial: Lei de Benford, voc√™ conhece?

A <b>Lei de Benford</b> √© um m√©todo utilizado amplamente para se <u>verificar fraudes</u> em conjuntos de dados!<br><br>
Essa lei estat√≠stica fala sobre a <u>distribui√ß√£o das frequ√™ncias dos d√≠gitos</u> e pode ser utilizada em v√°rios datasets da vida real, incluindo valores de a√ß√µes, n√∫mero de popula√ß√µes, imposto de rendas e entre outros...
<br><br>
Bom, como sou curioso! Resolvi usar ela para correlacionar a distribui√ß√£o dos n√∫meros do primeiro d√≠gito dos empenhos com a distribui√ß√£o refer√™ncia da Lei de Benford.<br><br>

Vamos ver no que vai dar? üòÜ

<b>Recomendo a leitura:</b><br> 
[TCU - Aplica√ß√µes da Lei de Benford √† auditoria de obras p√∫blicas](https://portal.tcu.gov.br/inovatcu/noticias/aplicacoes-da-lei-de-benford-a-auditoria-de-obras-publicas.htm)<br>

```python
# Criando colunas com os par√¢metros para utilizar a biblioteca da lei de Benford
df['p_r'] = df.N_EMPENHO/df.N_EMPENHO.shift()-1
df['l_r'] = np.log(df.N_EMPENHO/df.N_EMPENHO.shift())
df.tail()
```

Ao gerar o gr√°fico correlacionado os dados de empenho da Prefeitura com a lei de Benford voc√™ tem a possibilidade de apontar um √≠ndices de confiabilidade entre 0 a 100%. Para o gr√°fico a seguir, eu escolhi 95% de confiabilidade.

```python
# Setando os par√¢metros para plotar o gr√°fico correlacionando benford e os dados de empenho.
f1d = bf.first_digits(df.l_r, digs=1, decimals=8, confidence=95)
```

Para a an√°lise do <u>primeiro d√≠gito e com um √≠ndice de confiabilidade a 95%</u> <b>apenas</b> os d√≠gitos 3 e 4 corresponderam a lei. üòû

Levando em considera√ß√£o que barras em amarelo nos alerta que aquele d√≠gitos est√° fora da lei de Benford, podemos dizer que, nesse caso, o primeiro d√≠gito est√° <b>78%</b> sem corresponder a lei.

<h2>E se a gente fizer uma an√°lise com o segundo?</h2>

```python
sd = bf.second_digit(df.l_r, decimals=8, confidence=95)
```

Interessante! Com o segundo d√≠gito e √≠ndice de confiabilidade a 95% o conjunto de dados j√° se apresenta muito mais confi√°vel que o primeiro d√≠gito.

N√£o sei voc√™, mas minha curiosidade √© grande... E agora te pergunto...

<h2>E se a gente fizer a mesma an√°lise com os dois √∫ltimos n√∫meros?</h2>

```python
l2d = bf.last_two_digits(df.l_r, decimals=8, confidence=95)
```

Incr√≠vel, como apenas o primeiro d√≠gito apresentou uma alta porcentagem fora da lei de Benford.

A mantissa √© a parte decimal de um logaritmo.

Em um conjunto de dados de Benford, as mantissas dos registros s√£o uniformemente distribu√≠das, de modo que, quando solicitadas, <u>elas devem formar uma linha reta no intervalo</u>.
